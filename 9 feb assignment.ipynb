```python
#Ans1:
class vehicle:
    def __init__(self,name_of_vehicle,max_speed,average_of_vehicle):
        self.name_of_vehicle=name_of_vehicle
        self.max_speed=max_speed
        self.average_of_vehicle=average_of_vehicle
        
```

```python
#Ans 2:
class car(vehicle):
    def seating_capacity(self, capacity):
        return self.name_of_vehicle,self.capacity
```

```python
#Ans3: Multiple inheritance is when more then one class inherit property of a single class then it is called as multiple inheritance
#for example:
class school:
    def __init__ (self,profile,name,date_of_join):
        self.profile=profile
        self.name=name
        self.date_of_join=date_of_join

class student(school):
    pass

class teacher(school):
    pass

# As given in above example student and teacher are both child class of school class this is called multiple inheritance.
```

```python
#Ans 4: Getter is the used to access the private variable in a class. while setter is used to set the private variable.
# for example:
class getset:
    def __init__ (self,name,phone):
        self.name=name
        self.__phone=phone
    @property
    def getter(self):
        return self.__phone
    @getter.setter
    def setter(self,number):
        self.__phone=number
```

```python
#Ans5: Method overriding means to define the method in each class in subclasses and making it different from the parent class
#for example:
class dog:
    def legs(self):
        pass
    def sound(self):
        print("bark")
    def ear(self,ears):
        self.ears=ears
        print(ears)
class cow(dog):
    def sound(self):
        print("maow")
```

```python
bull_dog=dog()
jersey=cow()
```

```python
bull_dog.sound()
jersey.sound()
```
